**Label Encoding**

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['experience_level'] = le.fit_transform(df['experience_level'])
df['employment_type'] = le.fit_transform(df['employment_type'])
df['job_title'] = le.fit_transform(df['job_title'])
df['salary_currency'] = le.fit_transform(df['salary_currency'])
df['salary_in_usd'] = le.fit_transform(df['salary_in_usd'])
df['employee_residence'] = le.fit_transform(df['employee_residence'])
df['company_location'] = le.fit_transform(df['company_location'])
df['company_size'] = le.fit_transform(df['company_size'])
```

**Correlation**

```python
corr = df.corr()
#Creating a heatmap 
import seaborn as sn
plt.figure(figsize=(10,5))
sn.heatmap(corr, annot = True , fmt = ".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
X=df.drop(['salary_in_usd'], axis = 1)
Y=df["salary_in_usd"]
```

**Splitting The Dtaset**

```python
# Import library
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

**Modelling**

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import AdaBoostRegressor
```

**RandomForestRegressor()**

```python
model= RandomForestRegressor()
```

```python
model.fit(X_train,Y_train)
```

```python
y_predict = model.predict(X_test)
print("Random Forest  Regressor accurracy score")
model.score(X_train,Y_train)
```

**AdaBoostRegressor**

```python
model= AdaBoostRegressor()
```

```python
model.fit(X_train,Y_train)
```

```python
y_predict = model.predict(X_test)
```

```python
model.score(X_test,Y_test)
```

**DecisionTreeRegressor**

```python
model= DecisionTreeRegressor()
```

```python
model.fit(X_train,Y_train)
```

```python
y_predicted = model.predict(X_test)
```

```python
model.score(X_test,Y_test)
```

**KNeighborsRegressor**

```python
model=KNeighborsRegressor()
model.fit(X_train,Y_train)
```

```python
y_predicted=model.predict(X_test)
```
